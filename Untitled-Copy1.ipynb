```python
class Stock():
  def __init__(self, symbol, name, previousClosingPrice, currentPrice):
    self.__symbol = symbol
    self.__name = name
    self.__previousClosingPrice = previousClosingPrice
    self.__currentPrice = currentPrice

  def getName(self):
    return self.__name

  def getStock(self):
    return self.__symbol

  def getPPrice(self):
    return self.__previousClosingPrice 

  def setPPrice(self):
    return self.__currentPrice

  def setCPrice(self, previousClosingPrice):
    self.__previousClosingPrice = previousClosingPrice

  def setCPrice(self, currentPrice):
    self.__currentPrice = currentPrice

  def getChangePercent(self):
    cPrice = self.__previousClosingPrice
    pCPrice = self.__currentPrice
    changePercent = (cPrice - pCPrice) / pCPrice *100 
    return changePercent



def main():
    intelStock = Stock('INTC', 'Intel Corporation', 20.5, 20.35)
    print(f'The Percentage change is: {intelStock.getChangePercent(): .2%}')
    
main()
```
